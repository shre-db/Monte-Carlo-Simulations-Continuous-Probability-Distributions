```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
# Monte Carlo Simulation of Normal Distribution

mu = 5  # Mean
sigma = 2  # Standard Deviation

N = 10**6  # Number of random samples to generate

# Generate random numbers from the normal distribution
random_data = np.random.normal(mu, sigma, N)

plt.hist(random_data, bins=100, density=True, alpha=0.6, color='lightgreen', label='Simulated Data')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Monte Carlo Simulation of Normal Distribution')
plt.legend()
plt.grid(alpha=0.1)
plt.show()

# Calculate and print some statistics
mean_est = np.mean(random_data)
std_dev_est = np.std(random_data)
print("Estimated Mean: {}".format(mean_est))
print("Estimated Standard Deviation: {}".format(std_dev_est))
```

```python
# Monte Carlo Simulation of Continuous Uniform Distribution

from numpy.random import uniform as U

# Number of random samples to generate
N = 10**5

# Interval bounds
lower_bound = 20
upper_bound = 40


# Monte Carlo Simulation
numbers = []
for _ in range(N):
    # Generate a random number from the uniform distribution [20, 40]
    random_number = U(lower_bound, upper_bound)
    numbers.append(random_number)

plt.hist(numbers, bins=10, density=True, alpha=0.6, color='lightgreen', label='Simulated Data')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Monte Carlo Simulation of Uniform Distribution')
plt.legend()
plt.grid(alpha=0.1)
plt.show()

hist_values, _ = np.histogram(numbers, bins=10, density=True)
print(f"Estimated Probability: {np.mean(hist_values)}")
```

```python
# Monte Carlo Simulation of Exponential Distribution in Python.

# Parameters for the exponential distribution
lam_a = 0.1
lam_b = 0.2

# Number of random samples to generate
N = 10**5

# Generate random numbers from the Exponential distribution
random_data_a = np.random.exponential(scale=1/lam_a, size=N)
random_data_b = np.random.exponential(scale=1/lam_b, size=N)

# Plot the histogram to visualize the distribution
plt.hist(random_data_a, bins=50, density=True, alpha=0.4, color='lightgreen', label='Simulated A')
plt.hist(random_data_b, bins=50, density=True, alpha=0.4, color='lightblue', label='Simulated B')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title("Monte Carlo Simulation of Exponential Distribution")
plt.legend()
plt.grid(alpha=0.1)
plt.show()
```

```python
# Monte Carlo Simulation of Log Normal Distribution

mean_log = 0.0
std_log = 0.5

# Number of random samples to generate
N = 10**5

# Generate random numbers from the log-normal distribution
log_normal_data = np.random.lognormal(mean_log, std_log, N)

# Plot a histogram to visualize the distribution
plt.hist(log_normal_data, bins=50, density=True, alpha=0.6, color='lightgreen', label='simulated Data')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title("Monte Carlo Simulation of Log-Normal Distribution")
plt.grid(alpha=0.1)
plt.show()
```

```python
# Monte Carlo Simulation of Chi-Squared Distribution

# Degrees of Freedom
dof_a = 5
dof_b = 7

# Number of Samples
N = 10000

# Generate random numbers from the Chi-Squared Distribution
chi_a = np.random.chisquare(df=dof_a, size=N)
chi_b = np.random.chisquare(df=dof_b, size=N)


# Plot a histogram to visualize the distribution
plt.hist(chi_a, bins=50, density=True, alpha=0.6, color='lightgreen', label='Simulated A')
plt.hist(chi_b, bins=50, density=True, alpha=0.6, color='lightblue', label='Simulated B')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title("Monte Carlo Simulation of Chi-Squared Distribution")
plt.legend()
plt.grid(alpha=0.1)
plt.show()
```

```python
# Monte Carlo Simulation of Beta Distribution


# Parameters
alpha_x = 2
beta_x = 5

alpha_y = 4
beta_y = 5

# Number of random samples to generate
N = 10000

# Generate random nubers from the beta distribution
beta_data_x = np.random.beta(a=alpha_x, b=beta_x, size=N)
beta_data_y = np.random.beta(a=alpha_y, b=beta_y, size=N)

# Plot a histogram to visualize the distribution
plt.hist(beta_data_x, bins=50, density=True, alpha=0.6, color='lightgreen', label="Simulated X")
plt.hist(beta_data_y, bins=50, density=True, alpha=0.6, color='lightblue', label="Simulated Y")
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title("Monte Carlo Simulation of Beta Distribution")
plt.legend()
plt.grid(alpha=0.1)
plt.show()
```
